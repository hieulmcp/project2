# Step 2 - Recency, Frequency, Monetary Value analysis

### RFM analysis
- R: When is the latest purchase date? - Ngày mua gần nhất là ngày nào ?
- F: How frequently do they make purchases? - Họ mua hàng với tân suất như thế nào ?
- M: How large their average monetary value is made? - Giá trị tiền tệ trung bình của họ lớn như thế nào?
#### Giả định
- Chia nhóm theo giả định của bộ phân marketing
    + RFM được nhóm MKT nghiên cứu hành vi dữa trên 3 yếu tố như sau:
        - Recency (R) : Đo lường số ngày kể từ ngày mua hàng cuối cùng (lần truy cập gần đây nhất) đến ngày giả định chung để tính toán	
        - Frequency (F): Đo lường số lượng giao dịch (tổng số lần mua hàng) được thực hiện trong thời gian nghiên cứu	
        - Monetary Value (M): đo lường số tiền mà mỗi khách hàng đã chi tiêu trong thời gian nghiên cứu
    + Đánh giá khách hàng theo tiêu chí giả định như sau
        - Tiêu chí sếp loại khách hàng: Bạch kim, vàng, bạc 
        - Nếu điểm RFM lớn hơn hoặc bằng 10, cấp độ phải là "Bạch kim"
        - Nếu nó nằm trong khoảng từ 6 đến 10 thì nó phải là "Vàng" và nếu không thì nó phải là "Bạc", dưới 5 thì đồng
- Chia segment như sau:
    - Low value: Những khách hàng ít hoạt động hơn những khách hàng khác, không phải là khách thường xuyên và tạo ra doanh thu rất thấp (Doanh thu = 0 và có thể -)	
    - Mid value: Giá trung bình ở các chỉ tiêu. Nhóm khách hàng đông, thường xuyên, tạo ra doanh thu vừa phải	
    - High value: Nhóm khách hàng đep lại doanh thu cao, tần suất mua hàng thường xuyên. Nhóm khách hàng doanh nghiệp muốn giữ chân nhiều nhất	
### Mục tiêu kế hoạch marketing cần làm
- Thực hiện cách tiếp thị khách hàng dựa trên phân loại khách hàng để làm gì ?
    + Giành được khách hàng với tốn kém gấp 5-10 lần nhiêu hơn so với việc thỏa mãn và giữ chân khách hàng
    + Theo phân tích clohort analysis => Trung bình mỗi năm thì lượng mất khách hàng khoảng 50% lượng khách hàng hiện có
    + Giảm 5% tỷ lệ mất khách hàng này thì có thể làm tăng 25% lợi nhuận, tuy vào việc kinh doanh
    + Làm sao tăng tỷ lệ lợi nhuận trên khách hàng có xu hướng tăng trong suốt chu kỳ sống của khách hàng mà công ty giữ được
- Database MKT tiếp thị trên cơ sở dữ liệu
    + Dữ liệu quan trọng nhất cần ghi nhận là lịch sử mua hàng của từng khách hàng:
        + Biết được trước đây khách hàng đã từng mua gì cho ta manh mối để xem họ có thể sẽ thích mua gì cho lần tiếp theo
        + Nếu có thêm những thông tin dân số học về mỗi người mua: tuổi tác, học vấn, thu nhập, số người trong gia đình và những thuộc tính khác nếu có/ Đối với người mua là doanh nhân, đó là chức vụ công việc, nhiệm vụ công tác, quan hệ nghề nghiệp và địa chỉ liên lạc
        + Bản có thể bổ sung tâm lý thông tin về tâm lý mô tả những hoạt động, mối quan tâm và quan điểm (AIO - Activies, interests, opinions) của cá nhân khách hàng và suy nghĩ, quyết định, và ảnh hưởng đến người khác ra sao.
- Điều công ty chung tôi mong muốn
    + Chúng tôi muốn biết nhiều điều về mỗi khách hàng và khách hàng tiềm năng
    + Chúng tôi muốn lôi kéo khách hàng bằng việc chào hàng đúng đối tượng khách hàng, cả những mói hàng họ chưa biết hay lúc đầu chưa quan tâm
    + Chúng tôi muốn tiếp cận khách hàng bằng bất kỳ phương tiện truyền thông nào miễn là tiết kiệm nhưng hiệu quả, không cần biết họ có thích hay không
    + Chúng tôi muốn cắt giảm chi phí nói chuyện trực tiếp với khách hàng qua điện thoại
- Đưa ra những kế hoạch cho những nhòm khách hàng như sau:
    + Khách hàng "Bạch kim": Thực hiện giữ chân khách hàng và thực hiện các chương trình như tri ân khách hàng
    + Khách hàng "Vàng": Thêm các chương trình để kéo lên cho đạt chỉ tiêu điểm của doanh thu và tần suất mua hàng mong muốn
        + Nhân viên tiếp thị: từ xa có thể gọi điện cho khách hàng, gặp gỡ và đưa ra những chương trình khuyến mãi để tăng phần dịch vụ của công ty tốt hơn
    + Khách hàng "Bạc": Thì sẽ phân loại lại xem khách hàng nào khách cần thiết để khuyển mãi riêng và tạo ấn tượng để kéo lên thành viên vàng


## A. Import thư viện

```python
import numpy as np
import pandas as pd
import datetime as dt
```

## B. UPLOAD DATA

```python
# 2. Upload data => name file: House_data.xlsx
dir_file = "data_analysis/OnlineRetail.csv"
df = pd.read_csv(dir_file, encoding='latin1')

# convert object to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# convert object to float
#df['UnitPrice'] = df['UnitPrice'].apply(lambda x: x.replace(',', '.'))
df['UnitPrice'] = df['UnitPrice'].apply(lambda col:pd.to_numeric(col, errors='coerce'))
```

### 1.1. Lọc dữ liệu
- Khi phân tích bài toán RFM thì cần quan tâm đến thời gian: tròn đủ 12 tháng

```python
df = df.loc[(df['InvoiceDate'] >= '2010-12-10') & (df['InvoiceDate'] < '2011-12-13')]
```

```python
df.head()
```

```python
print('Min:{}; Max:{}'.format(min(df.InvoiceDate),
                              max(df.InvoiceDate)))
```

### 1.2. Xây dựng chuỗi thời gian trong bài toán RFM

```python
# Thêm 1 ngày để chúng ta có thể phân tích 1 ngày gần đây
snapshot_date = max(df.InvoiceDate) + dt.timedelta(days=1)
```

### 1.3. Tem thuộc tính
revenue = UnitPrice * Quantity

```python
df['revenue'] = df['UnitPrice'] * df['Quantity']
```

### 1.3. Xây dựng bộ dữ liệu chuẩn cho bài toán
- Tính toán ngày đăng nhập gần đây nhất
- Số hóa đơn 
- Tổng giá trị

```python
# Trung bình của mỗi khách hàng 
datamart = df.groupby(['CustomerID']).agg({
        'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
        'InvoiceNo': 'count',
        'revenue': 'sum'})

```

```python
# Rename columns for easier interpretation
datamart.rename(columns = {'InvoiceDate': 'Recency',
        'InvoiceNo': 'Frequency',
        'revenue': 'MonetaryValue'}, inplace=True)

```

```python
# Check the first rows
datamart.head()
```

### 1.4. Gán nhãn dữ liệu theo tứ phân vị

```python
# Gán nhãn cho tần suất đăng nhập, tần xuất giao dịch
r_labels = range(3, 0, -1); f_labels = range(1, 4)
```

```python
# Gán nhãn cho các nhóm phân vị bằng nhau
r_groups = pd.qcut(datamart['Recency'], q=3, labels=r_labels)
```

```python
# Gán nhãn cho các nhóm phân vị bằng nhau
f_groups = pd.qcut(datamart['Frequency'], q=3, labels=f_labels)
```

```python
# Thêm cột gãn nhãn
datamart = datamart.assign(R=r_groups.values, F=f_groups.values)
```

```python
datamart
```

### 1.5. Tính điểm RFM
- Thêm nhãn cho cột MonetaryValue
- Gán nhãn cho cột trên theo số
- Thêm cột M
- Tính toán RFM_Score

```python
# Thêm nhãn cho cột MonetaryValue
m_labels = range(1, 4)
```

```python
# Thêm nhãn bằng nhau cho cột MonetaryValue
m_groups = pd.qcut(datamart['MonetaryValue'], q=3, labels=m_labels)
```

```python
# Thêm nhãn cho cột M
datamart = datamart.assign(M=m_groups.values)
```

```python
# Tính RFM_Score
datamart['RFM_Score'] = datamart[['R','F','M']].sum(axis=1)
print(datamart['RFM_Score'].head())
```

### 1.6. Đánh giá khách hàng theo tiêu chí giả định như sau
- Tạo các phân đoạn có tên Trên cùng, Trung bình, Thấp
- Nếu điểm RFM lớn hơn hoặc bằng 10, cấp độ phải là "Top"
- Nếu nó nằm trong khoảng từ 6 đến 10 thì nó phải là "Trung bình" và nếu không thì nó phải là "Thấp"

```python
# Xác định phân cấp 
def rfm_level(df):
    if df['RFM_Score'] >= 10:
        return 'Bạch kim'
    elif (df['RFM_Score'] >= 7) and (df['RFM_Score'] < 10):
        return 'Vàng'
    elif (df['RFM_Score'] >= 5) and (df['RFM_Score'] < 7):
        return 'Bạc'
    else:
        return 'Đồng'
```

```python
# Thêm 1 cột phân cấp
datamart['RFM_Level'] = datamart.apply(rfm_level, axis=1)
```

```python
datamart.head()
```

### 1.7. Tính toán giá trị trung bình cho mỗi phân cấp - level và trả về mỗi nhóm
- Thứ nhất không có làm việc bộ phận khảo sát thị trường
- Thứ 2 thang điểm dựa trên kinh nghiệm bản thân để làm

```python
rfm_level_agg = datamart.groupby('RFM_Level').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'MonetaryValue': ['mean', 'count']
}).round(0)
rfm_level_agg.columns = rfm_level_agg.columns.droplevel()
rfm_level_agg.columns = ['RecencyMean','FrequencyMean','MonetaryMean','Count']
rfm_level_agg['Percent'] = round((rfm_level_agg['Count']/rfm_level_agg.Count.sum())*100,2)
rfm_level_agg['Percent_Recency'] = round((rfm_level_agg['RecencyMean']/rfm_level_agg.Count.sum())*100,2)
rfm_level_agg['Percent_Frequency'] = round((rfm_level_agg['FrequencyMean']/rfm_level_agg.Count.sum())*100,2)
rfm_level_agg['Percent_MonetaryValue'] = round((rfm_level_agg['MonetaryMean']/rfm_level_agg.Count.sum())*100,2)
```

```python
# Reset the index
rfm_level_agg = rfm_level_agg.reset_index()
rfm_level_agg
```
